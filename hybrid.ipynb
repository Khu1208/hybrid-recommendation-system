```python
comment_counts = pd.DataFrame(df["title"].value_counts())
comment_counts.head()
```

```python
common_movies=comment_counts[comment_counts["title"]>=100].index
```

```python
common_movies_df=df[df["title"].isin(common_movies)]
```

```python
user_movie_df = common_movies_df.pivot_table(index=["title"], columns=["userId"], values=["rating"])
user_movie_df.head()
```

```python
corr_mat=user_movie_df.T.corr()
corr_mat.head()
```

```python
movie_name = "Matrix, The (1999)"

recommemnd_movies=corr_mat[corr_mat.index==movie_name].unstack().sort_values(ascending=False)[0:11].index
for i in recommemnd_movies:
  print(i[0])
```

```python
random_user=2
```

```python
user_movie = common_movies_df.pivot_table(index=["userId"], columns=["title"], values=["rating"])
user_movie.head()
```

```python
random_user_df = user_movie[user_movie.index == random_user]
random_user_df
```

```python
movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
movies_watched
```

```python
len(movies_watched)
```

```python

movies_watched_df = user_movie[movies_watched]
movies_watched_df.head()
movies_watched_df.shape
```

```python
user_movie_count = movies_watched_df.T.notnull().sum()

user_movie_count = user_movie_count.reset_index()
user_movie_count.columns = ["userId","movie_count"]
user_movie_count.head()
```

```python
users_same_movies = user_movie_count[user_movie_count["movie_count"] > 0.4*len(movies_watched)]["userId"]
users_same_movies.count()
```

```python

final_df = movies_watched_df[movies_watched_df.index.isin(users_same_movies)]

final_df.head()
```

```python
final_df.T.corr()
```

```python
corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.index.names = ['user_id_1', 'user_id_2']
corr_df = corr_df.reset_index()
corr_df.head()
```

```python
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.45)][
    ["user_id_2", "corr"]].reset_index(drop=True)
top_users.shape
```

```python
top_users = top_users.sort_values(by='corr', ascending=False)
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
```

```python

top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how='inner')

top_users_ratings = top_users_ratings[top_users_ratings["userId"] != random_user]
top_users_ratings.shape
```

```python
top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})

recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()

recommendation_df.shape
```

```python
recommendation_df[recommendation_df["weighted_rating"] > 3]
```

```python
movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3.5].sort_values("weighted_rating", ascending=False)

movies_to_be_recommend=movies_to_be_recommend.merge(movie[["movieId", "title"]])
movies_to_be_recommend["title"].to_list()
```

```python


def recommend(user):

  movie_id = rating[(rating["userId"] == user) & (rating["rating"] == 5.0)].sort_values(by="timestamp", ascending=False)["movieId"][0:1].values[0]
  movie_one=movie[movie["movieId"]==movie_id]["title"].values[0]

  common_movies=comment_counts[comment_counts["title"]>=100].index
  new_df=df[(df["title"].isin(common_movies)) | (df["title"]==movie_one)]
  user_movie_df =new_df.pivot_table(index=["title"], columns=["userId"], values=["rating"])
  corr_mat=user_movie_df.T.corr()
  recommemnd_movies=corr_mat[corr_mat.index==movie_one].unstack(level=1).sort_values(ascending=False)[1:11].index

  user_movie = new_df.pivot_table(index=["userId"], columns=["title"], values=["rating"])
  random_user_df = user_movie[user_movie.index == random_user]
  movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
  movies_watched_df = user_movie[movies_watched]
  user_movie_count = movies_watched_df.T.notnull().sum()
  user_movie_count = user_movie_count.reset_index()
  user_movie_count.columns = ["userId","movie_count"]
  users_same_movies = user_movie_count[user_movie_count["movie_count"] > 0.4*len(movies_watched)]["userId"]
  final_df = movies_watched_df[movies_watched_df.index.isin(users_same_movies)]
  corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
  corr_df = pd.DataFrame(corr_df, columns=["corr"])
  corr_df.index.names = ['user_id_1', 'user_id_2']
  corr_df = corr_df.reset_index()
  top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.45)][
    ["user_id_2", "corr"]].reset_index(drop=True)
  top_users = top_users.sort_values(by='corr', ascending=False)
  top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
  top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how='inner')
  top_users_ratings = top_users_ratings[top_users_ratings["userId"] != random_user]
  top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
  top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
  recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
  recommendation_df = recommendation_df.reset_index()
  movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3.5].sort_values("weighted_rating", ascending=False)
  movies_to_be_recommend=movies_to_be_recommend.merge(movie[["movieId", "title"]])
  movies_to_be_recommend=movies_to_be_recommend["title"].to_list()
  for i in range(10):
   print(movies_to_be_recommend[i])
  for i in recommemnd_movies:
    print(i[0])


```
